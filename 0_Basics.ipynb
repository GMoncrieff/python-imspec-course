# Python core data types

## Arithmetic operations

Python supports the usual arithmetic operators: + (addition), * (multiplication), / (division), ** (power), // (integer division).

## Lists

Lists are a container type for ordered sequences of elements. They are declared inside square brackets `[]`. Lists can be initialized empty

```python
my_list = []
```

or with some initial elements

```python
my_list = [1, 2, 3]
```

Lists have a dynamic size and elements can be added (appended) to them

```python
my_list.append(4)
my_list
```

We can access individual elements of a list (indexing starts from 0)

```python
my_list[2]
```

We can access "slices" of a list using `my_list[i:j]` where `i` is the start of the slice (again, indexing starts from 0) and `j` the end of the slice. For instance:

```python
my_list[0:2]
```

Omitting the second index means that the slice shoud run until the end of the list

```python
my_list[1:]
```

We can check if an element is in the list using `in`

```python
5 in my_list
```

The length of a list can be obtained using the `len` function

```python
len(my_list)
```

## Dictionaries


In Python, a dictionary is a built-in data structure that stores a collection of key-value pairs. Dictionaries are indexed by unique keys. Keys and values can be strings or numeric data types. Dictionaries are declared in curly brackets `{}`

Here's a basic example of a dictionary in Python:

## Conditionals

As their name indicates, conditionals are a way to execute code depending on whether a condition is True or False. As in other languages, Python supports `if` and `else` but `else if` is contracted into `elif`, as the example below demonstrates.

```python
my_variable = 5
if my_variable < 0:
  print("negative")
elif my_variable == 0:
  print("null")
else: # my_variable > 0
  print("positive")
```

Here `<` and `>` are the strict `less` and `greater than` operators, while `==` is the equality operator (not to be confused with `=`, the variable assignment operator). The operators `<=` and `>=` can be used for less (resp. greater) than or equal comparisons.

Contrary to other languages, blocks of code are delimited using indentation. Here, we use 2-space indentation but many programmers also use 4-space indentation. Any one is fine as long as you are consistent throughout your code.

## Loops

Loops are a way to execute a block of code multiple times. There are two main types of loops: while loops and for loops.

While loop

```python
i = 0
while i < len(my_list):
  print(my_list[i])
  i += 1 # equivalent to i = i + 1
```

For loop

```python
for i in range(len(my_list)):
  print(my_list[i])
```

If the goal is simply to iterate over a list, we can do so directly as follows

```python
for element in my_list:
  print(element)
```

## Functions

To improve code readability, it is common to separate the code into different blocks, responsible for performing precise actions: functions. A function takes some inputs and process them to return some outputs.

```python
def square(x):
  return x ** 2

def multiply(a, b):
  return a * b

# Functions can be composed.
square(multiply(3, 2))
```

```python
square(multiply(a=3, b=2))
```

## Classes and Object Oriented Programming

Something that you may not have encountered yet is the concept of Object Oriented Programming. Object-Oriented Programming (OOP) is a way to write programs using "objects", which bundle related data and operations.

**Classes:** Think of them as blueprints for creating similar objects. An exmaple might be a Person class or a Dog class

**Objects:** These are individual instances of a class. They have specific data (like a person's name) and can perform operations (like a person walking). An example would be Glenn - an object of the Person class, or Murray - an object of the dog class

In OOP, you create instances from classes, similar to making houses from blueprints. When we define the class we can give it attributes and methods.

Let see what this looks like in concete term by defing a dog class

```python
class Dog:
    def __init__(self, name):
        self.name = name  # set an attribute using self.attribute

    def bark(self):     # create a method using def
        return "Woof!"
```

```python

# Now, let's create an instance of the class Dog
my_dog = Dog("Murray")
```

The `my_dog` instance has a single attribute - his name, Murray (a fine name for a dog). We can access an instances attributes using the pattern `object.attribute`

```python
my_dog.name
```

`my_dog` has a single method() - `bark()`. We can access a class' method using the pattern `instance.method`. Methods and functions are the same thing. A method is just a function that belongs to an instance

```python
my_dog.bark()
```

# NumPy

NumPy is a popular library for storing arrays of numbers and performing computations on them. Not only this enables to write often more succint code, this also makes the code faster, since most NumPy routines are implemented in C for speed.

Numpy is probably the most important package for doing data science and scientific analysis in Python. It is very powerful and is often the basis for other packages that add additional complexity (it is at the core of all raster handling packages in python)

To use NumPy in your program, you need to import it as follows

```python
import numpy as np
```

## Array creation



NumPy arrays can be created from Python lists

```python
my_array = np.array([1, 2, 3])
my_array
```

NumPy supports array of arbitrary dimension. For example, we can create two-dimensional arrays (e.g. to store a matrix) as follows

```python
my_2d_array = np.array([[1, 2, 3], [4, 5, 6]])
my_2d_array
```

We can access individual elements of a 2d-array using two indices

```python
# index using:
# position_in_dim1, position_in_dim2
my_2d_array[1, 2]
```

We can also access rows

```python
my_2d_array[1]
```

and columns

```python
my_2d_array[:, 2]
```

a bare index of `:` means give me everything in that dim

```python
print(my_2d_array[:, :])
#is the same as
print(my_2d_array)
```

Arrays have a `shape` attribute

```python
print(my_array.shape)
print(my_2d_array.shape)
```

Contrary to Python lists, NumPy arrays must have a type and all elements of the array must have the same type.

```python
my_array.dtype
```

The main types are `int32` (32-bit integers), `int64` (64-bit integers), `float32` (32-bit real values) and `float64` (64-bit real values).

The `dtype` can be specified when creating the array

```python
my_array = np.array([1, 2, 3], dtype=np.float64)
my_array.dtype
```

We can create arrays of all zeros using

```python
zero_array = np.zeros((2, 3))
zero_array
```

and similarly for all ones using `ones` instead of `zeros`.

We can create a range of values using

```python
np.arange(5)
```

or specifying the starting point

```python
np.arange(3, 5)
```

Another important operation is `reshape`, for changing the shape of an array

```python
my_array = np.array([1, 2, 3, 4, 5, 6])
my_array.reshape(3, 2)
```

Play with these operations and make sure you understand them well.

## Basic operations

In NumPy, we express computations directly over arrays. This makes the code much more succint.

Arithmetic operations can be performed directly over arrays. For instance, assuming two arrays have a compatible shape, we can add them as follows

```python
array_a = np.array([1, 2, 3])
array_b = np.array([4, 5, 6])
array_a + array_b
```

NumPy has a range of functions that can be applied to arrays, e.g. `np.sin`

```python
np.sin(array_a)
```

Vector inner product can be performed using `np.dot`

```python
np.dot(array_a, array_b)
```

numpy arrays also have their own methods e.g matrix transpose `.transpose()`

```python
array_a.transpose()
```

## Slicing and masking

Like Python lists, NumPy arrays support slicing

```python
z = np.arange(10)
# you can read this as:
# start at the 5th postion and give me everything from there until the end of that dim
print(z[5:])
# you can read this as:
# start at the beggining and give me everything until the 5th position
print(z[:5])
```

We can also select only certain elements from the array

```python
x = np.arange(10)
mask = x >= 5
x[mask]
```

Because numpy arrays can have many dimensions, indexing can get quite complicated. If you want to know more have a look at the docs [here](https://numpy.org/doc/stable/user/basics.indexing.html). Lets look at what this looks like on a 3D array

```python
# Create a 3D array with values ranging from 0 to 269 with shape 3x3x30.
array_3d = np.arange(270).reshape(3, 3, 30)
array_3d
```

Lets pretend that this array has dimension latitude, longitude and band (3x3x30). Now lets extract a spatial subset  - we onyl want the first 2x2 lat long grid with all wavelengths

```python
#dimensions separated by comma
#1st dim - :2 indicates we want everything up to the 2nd coordinate i.e 0 and 1. Writing 0:2 would be equivalent
#2nd dim - sanme as above
#3rd dim -  : a bare semicolon indicate we want all entires in that dimension
sliced_array = array_3d[:2, :2, :]
sliced_array
```

To confirm our array has the correct shape (we expect 2x2x300 we can check the array's `shape` attribute

```python
sliced_array.shape
```

# Pandas


Pandas is the Python library used for tabular data manipulation and analysis. It is **very** widely used. It is the Python version of dplyr, tidyr and readr rolled into one.

Here's an example of loading a DataFrame from a CSV file:

```python
import pandas as pd

# Load data from a CSV file into a DataFrame
df = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv')

# Display the first 5 rows of the DataFrame
print(df.head())
```

There are lots and lots of tutorial on how to wrnagle data with pandas. I will just show you some basics

####Indexing
Or how to select particular rows and columns

```python
# Accessing a column
print(df['sepal_length'])

# use .loc to access a row by label
print(df.loc[0])

# use .iloc to access a row by integer index
print(df.iloc[0])
```

#### Filtering
or selecting rows by a logical condition


```python
# Filter rows where 'sepal_length' is greater than 5
filtered = df[df['sepal_length'] > 5]

#what is happening here is that df['sepal_length'] > 5 is creating a series of True/False values based on the condition
#this is then used to retrun rows where the series is True

print(filtered)
```

another way of achieving the same thing is using `query`

```python
filtered = df.query('sepal_length > 5')

print(filtered)
```

#### Grouping and aggregation

```python
# Group data by the values in 'species' and compute the mean of the other columns for each group
grouped = df.groupby('species').mean()
print(grouped)
```

```python
# or maybe we just want the aggregation of a single colmn
petal_grouped = df.groupby('species')['petal_length'].mean()
print(petal_grouped)
```

#### Chaining functions
We can chain things together if we enclose the entire chain in brackets (like the `%>%` or `|>` R pipe operator). You need to think of this a little different from the R operator.   
In R, the pipe operator passes the results coming from the left side as the first argument to the function on it's right side.   
In Python, there is no explicit pipe, but we are able to call all of the methods of an instance using `.` even if we have not assigned it yet

```python
#first assign the result
result_temp = df.groupby('species')
#then call its mean() method
result = result_temp.mean()
```

is the same as

```python
#we can use the .mean() method without explicitly assinging the result of the preceding function
#the trick is that you have to know the the type/class of the first function is to know what methods it has
#but with pandas the results are usually another dataframe
result = df.groupby('species').mean()
```

Lets see how we could use this to chain many things together

```python
#lets do everything above in one step
result = (pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv')
          .query('sepal_length > 5')
          .groupby('species')
          .mean())

print(result)

```

In contrast to R where everyone now uses ggplot, there are a few different plotting libraries in python. ### Matplotlib

#### Basic plots

We start with a rudimentary plotting example.

```python
from matplotlib import pyplot as plt

#create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

#plt is a class for plots
#it has methods like 'plot()', or '.show()' that modifiy it

#create the figure
plt.figure()
#add a plot
plt.plot(x, y1, label="Sinusoid")
#add some styling
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Matplotlib example")
plt.legend(loc="upper left")
#show the plot
plt.show()
```

Adding multiple lines in simple

```python

plt.figure()
plt.scatter(x, y1, color="blue")
plt.scatter(x, y2, color="green")
plt.show()
```

Yoiu might also see matplotlib plots expressed using the `fig, ax` syntax

```python
# create a new figure with a single Axes
fig, ax = plt.subplots()

#add data the the axes
ax.scatter(x, y1, label="Sinusoid1",color="blue")
ax.scatter(x, y2, label="Sinusoid2", color="green"
)
# Add a legend
ax.legend()

plt.show()
```

fig is the whole window/page in the plot. Think of it as a container that holds everything.

ax is the individual plot inside the Figure. It contains the elements of the plot (lines, points, labels, etc.), you can also put multiple axes in a single figure

When you use `fig, ax = plt.subplots()`, you're creating a new Figure (fig) and a new plot (ax) within that Figure. Using ax gives you fine control over each specific plot, which is useful in complex plotting scenarios.

```python
#multiple plots

# Create a figure and two axes in a single column
fig, axs = plt.subplots(2, 1)

# Plot on each set of axes
axs[0].plot(x, y1, 'r')  # red line
axs[0].set_title('Sine function')

axs[1].plot(x, y2, 'b')  # blue line
axs[1].set_title('Cosine function')

# Automatically adjust the subplot parameters for better layout
fig.tight_layout()

plt.show()
```
